```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.columns
```

```python
obj_cols = df.select_dtypes(include='object').columns
df[obj_cols] = df[obj_cols].apply(lambda x: x.str.strip().str.title())
```

```python
df.isnull().sum()
```

```python
df[['growth_rate_percent', 'employee_count', 'funding_stage']].dtypes
```

```python
# Using Median to fill:
df['growth_rate_percent'] = df['growth_rate_percent'].fillna(df['growth_rate_percent'].median())
```

```python
# Using Integer:
df['employee_count'] = df['employee_count'].fillna(int(df['employee_count'].median()))
```

```python
# Using Unknown Values:
df['funding_stage'] = df['funding_stage'].fillna('Unknown')
```

```python
# Using Median:
df['funding_amount_cr'] = df['funding_amount_cr'].fillna(df['funding_amount_cr'].median())
```

```python
# Using Median:
df['valuation_cr'] = df['valuation_cr'].fillna(df['valuation_cr'].median())
```

```python
df.isnull().sum()
```

```python
# STEP 1: FILTER TOP INDUSTRIES (TOP 10 BY FREQUENCY)
# We are keeping only the top 10 industries with the highest startup count.
top_industries = df['industry'].value_counts().head(10).index
df = df[df['industry'].isin(top_industries)]
```

```python
# STEP 2: FILTER DATA FOR RECENT YEARS (2018 ONWARDS)
# Keeping data from 2018 onwards gives a more relevant view of modern startups.
df = df[df['year'] >= 2018]
```

```python
# STEP 3: KEEPING ONLY VALID FUNDING AMOUNTS
# To maintain accuracy in visualizations, we keep only startups with funding greater than 0 crore.
df = df[df['funding_amount_cr'] > 0]
```

```python
# STEP 4: VERIFY FILTERING
print("Final Shape:", df.shape)
print("\nIndustries remaining:", df['industry'].unique())
print("\nYears range:", df['year'].min(), "to", df['year'].max())
```

```python
# Grouping total funding by industry
industry_funding = df.groupby('industry')['funding_amount_cr'].sum().reset_index()
# Sorting in descending order to find top funded industries
industry_funding = industry_funding.sort_values(by='funding_amount_cr', ascending=False)
```

```python
# Display top 10 industries
print(industry_funding.head(5))
```

```python
#  UNIVARIATE ANALYSIS
# Industry-wise Startup Count
sns.countplot(data=df, y='industry', order=df['industry'].value_counts().index, palette='Greens_r')
plt.title('Industry-wise Startup Count', fontsize=14, fontweight='bold')
plt.xlabel('Number of Startups', fontsize=12, fontweight='bold')
plt.ylabel('Industry', fontsize=12, fontweight='bold')
plt.show()
```

```python
# Funding Stage Distribution
sns.countplot(data=df, x='funding_stage', palette='Blues_r')
plt.title('Distribution of Funding Stages', fontsize=14, fontweight='bold')
plt.xlabel('Funding Stage', fontsize=12, fontweight='bold')
plt.ylabel('Count', fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Funding Amount (Cr) Distribution
sns.histplot(df['funding_amount_cr'], bins=30, kde=True, color='mediumseagreen')
plt.title('Funding Amount (Cr) Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Funding Amount (Cr)', fontsize=12, fontweight='bold')
plt.ylabel('Frequency', fontsize=12, fontweight='bold')
plt.show()
```

```python
# BIVARIATE ANALYSIS
# Get top 5 industries by funding
top5_industries = industry_funding.head(5)

plt.figure(figsize=(7,7))
plt.pie(top5_industries['funding_amount_cr'],
        labels=top5_industries['industry'],
        autopct='%1.1f%%',
        colors=sns.color_palette('pastel'),
        startangle=90,
        wedgeprops={'width':0.4})
plt.title('Funding Share by Top 5 Industries', fontsize=14, fontweight='bold')
plt.show()
```

```python
# Year vs Total Funding
yearly_funding = df.groupby('year')['funding_amount_cr'].sum().reset_index().sort_values(by='year')
sns.lineplot(data=yearly_funding, x='year', y='funding_amount_cr', marker='o', color='orange')
plt.title('Total Funding Trend by Year', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12, fontweight='bold')
plt.ylabel('Total Funding (Cr)', fontsize=12, fontweight='bold')
plt.show()
```

```python
# Funding Amount vs Valuation (by Stage)
sns.scatterplot(data=df, x='funding_amount_cr', y='valuation_cr', hue='funding_stage', palette='pastel')
plt.title('Funding vs Valuation by Funding Stage', fontsize=14, fontweight='bold')
plt.xlabel('Funding Amount (Cr)', fontsize=12, fontweight='bold')
plt.ylabel('Valuation (Cr)', fontsize=12, fontweight='bold')
plt.legend(title='Funding Stage')
plt.show()
```

```python
# CORRELATION HEATMAP
corr = df[['funding_amount_cr', 'valuation_cr', 'growth_rate_percent', 'employee_count']].corr()
sns.heatmap(corr, annot=True, cmap='Greens', fmt=".2f", linewidths=0.5)
plt.title('Correlation Between Key Numerical Columns', fontsize=14, fontweight='bold')
plt.show()
```
